```r
# Ajuste Lineal
datos <- read.csv("energia_eolica.csv")
x <- datos$velocidad.viento
y <- datos$energia
plot(x,y)
recta <- lm(y~x)
summary(recta)
abline(recta,col="blue")
text(10,900,"y = -235 + 12x",col="red")
```

```r
# Ajuste polinomio grado 2
x2 <- x^2
parabola <- lm(y~x+x2)
summary(parabola)
y2 <- 42.626494 -7.928610*x + 0.253518*x^2
plot(x, y)
lines(x,y2,col="green")
text(10,900,"y = 42 7.9x + 0.25 x2",col="green")
```

```r
# Ajuste polinomio grado 3
x3 <- x^3
cubica <- lm(y~x+x2+x3)
plot(x,y,col = "red",pch = 21,bg="blue")
coef(cubica)
y3 <- -20 + 1.99*x - 0.06*x^2 + 0.0023* x^3
points(x,y3,col="orange")
```

```r
# Graficas separadas
par(mfrow = c(1,3))
plot(x,y,las=1,main="Lineal",col.main="red")
abline(recta,col="red")
text(20,900,"y = -235 + 12x",col="red")
```

```r
datos <- read.csv("energia_eolica.csv")
x <- datos$velocidad.viento
y <- datos$energia
recta <- lm(y ~ x)
plot(x, y,
     main = "grafica lineal daniel",
     xlab = "Wind Speed",
     ylab = "Energy",
     col = "blue",
     pch = 16)
abline(recta, col = "red", lwd = 2)
text(10, 900, paste("y =", round(coef(recta)[1], 2), "+", round(coef(recta)[2], 2), "x"), col = "red")
```

```r
datos <- read.csv("energia_eolica.csv")
x <- datos$velocidad.viento
y <- datos$energia
x2 <- x^2
parabola <- lm(y ~ x + x2)
y2 <- predict(parabola, data.frame(x = x, x2 = x2))
plot(x, y,
     main = "Quadratic Regression of Wind Speed vs. Energy",  # Add title
     xlab = "Wind Speed",
     ylab = "Energy",
     col = "blue",  # Change data point color
     pch = 16)
lines(x, y2, col = "red", lwd = 2)
equation <- paste("y =", round(coef(parabola)[1], 2), "+",
                  round(coef(parabola)[2], 2), "x +",
                  round(coef(parabola)[3], 2), "x^2")
text(10, 900, equation, col = "red")
```

```r
datos <- read.csv("energia_eolica.csv")
x <- datos$velocidad.viento
y <- datos$energia
x2 <- x^2
x3 <- x^3
cubica <- lm(y ~ x + x2 + x3)
y3 <- predict(cubica, data.frame(x = x, x2 = x2, x3 = x3))
plot(x, y,
     main = "Cubic Regression of Wind Speed vs. Energy",  # Add title
     xlab = "Wind Speed",
     ylab = "Energy",
     col = "blue",  # Change data point color
     pch = 16)
equation <- paste("y =", round(coef(cubica)[1], 2), "+",
                  round(coef(cubica)[2], 2), "x +",
                  round(coef(cubica)[3], 2), "x^2 +",
                  round(coef(cubica)[4], 2), "x^3")
text(10, 900, equation, col = "red")
```
