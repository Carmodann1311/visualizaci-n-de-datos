```r
data(cars)
```

```r
summary(cars)
```

```r
View(cars)
```

```r
if(!require(ggplot2)) install.packages("ggplot2")

library(ggplot2)

ggplot(cars, aes(x = speed, y = dist)) +
  geom_point() +
  labs(title = "Gráfico de dispersión de Speed vs Dist",
       x = "Speed",
       y = "Dist")
```

```r
if(!require(ggplot2)) install.packages("ggplot2")

library(ggplot2)

ggplot(cars, aes(x = speed, y = dist)) +
  geom_line() +
  labs(title = "Gráfico de líneas de Speed vs Dist",
       x = "Speed",
       y = "Dist")
```

```r
if(!require(ggplot2)) install.packages("ggplot2")

library(ggplot2)

ggplot(cars, aes(x = speed)) +
  geom_histogram(fill = "blue") +
  labs(title = "Histograma de Speed",
       x = "Speed",
       y = "Frecuencia")
```

```r
ggplot(cars, aes(x = dist)) +
  geom_histogram(fill = "green") +
  labs(title = "Histograma de Dist",
       x = "Dist",
       y = "Frecuencia")
```

```r
ggplot(cars, aes(x = speed)) +
  geom_density(fill = "blue", alpha = 0.5) +
  labs(title = "Density Plot of Speed",
       x = "Speed",
       y = "Density")
```

```r
ggplot(cars, aes(x = dist)) +
  geom_density(fill = "green", alpha = 0.5) +
  labs(title = "Density Plot of Dist",
       x = "Dist",
       y = "Density")
```

```r
ggplot(cars, aes(y = speed)) +
  geom_boxplot(fill = "blue", alpha = 0.5) +
  labs(title = "Box Plot of Speed",
       y = "Speed")
```

```r
ggplot(cars, aes(y = dist)) +
  geom_boxplot(fill = "green", alpha = 0.5) +
  labs(title = "Box Plot of Dist",
       y = "Dist")
```

```r
cars$speed_factor <- as.factor(cars$speed)
```

```r
ggplot(cars, aes(x = "", fill = speed_factor)) +
  geom_bar(width = 1) +
  coord_polar("y", start = 0) +
  labs(title = "Pie Chart of Speed") +
  theme(axis.text.x = element_blank())
```

```r
cars$dist_factor <- as.factor(cars$dist)
```

```r
ggplot(cars, aes(x = "", fill = dist_factor)) +
  geom_bar(width = 1) +
  coord_polar("y", start = 0) +
  labs(title = "Pie Chart of Dist") +
  theme(axis.text.x = element_blank())
```

```r
cor(cars$speed, cars$dist, method = "pearson")
```

```r
cor(cars$speed, cars$dist, method = "kendall")
```

```r
cor(cars$speed, cars$dist, method = "spearman")
```
