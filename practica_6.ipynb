```r
#Correlacion
peso<-c(85,65,80,56,60,67,78,65,60,92)
est<-c(1.8,1.61,1.71,1.57,1.65,1.73,1.79,1.67,1.59,1.83)
den<-density(peso)
plot(den,col="red")
summary(peso)
boxplot(peso,col="yellow1")
points(mean(peso),pch=8)
```

```r
den<-density(est)
plot(den,col="red")
summary(est)
boxplot(est,col="yellow1")
points(mean(est),pch=8)
```

```r
plot(est,peso,col=rainbow(13:25),pch=18,main="Dispersion")
```

```r
str(peso)
class(peso)
```

```r
cor(est,peso)
cor(peso,est)
```

```r
cor(est,peso,method = "kendall")
cor(est,peso,method = "spearman")
```

```r
#Regresion lineal
mod <-lm(peso~est)
summary(mod)
coef(mod)
plot(est,peso,pch=21,bg="cyan")
abline(mod,col=3,lwd=3)
```
