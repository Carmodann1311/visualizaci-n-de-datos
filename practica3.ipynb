```r
# Crea un dataframe con los datos
data <- data.frame(
  V1 = c(97.0, 106.5, 111.0, 110.6, 115.0, 124.0),
  V2 = c(97.0, 107.5, 113.0, 111.8, 116.0, 128.0)
)
```

```r
# Cargar la librería necesaria
library(ggplot2)

# Crear la gráfica lineal
ggplot(data, aes(x = seq_along(V1))) +
  geom_line(aes(y = V1, color = "V1")) +
  geom_line(aes(y = V2, color = "V2")) +
  labs(x = "Índice", y = "Valores", color = "Variables") +
  theme_minimal() +
  ggtitle("Gráfica Lineal de V1 y V2")
```

```r
# Crear un dataframe con los datos
data <- data.frame(
  Variables = c("V1", "V2"),
  Valores = c(mean(c(97.0, 106.5, 111.0, 110.6, 115.0, 124.0)),
              mean(c(97.0, 107.5, 113.0, 111.8, 116.0, 128.0)))
)
```

```r
# Cargar la librería necesaria
library(ggplot2)

# Crear el gráfico de barras
ggplot(data, aes(x = Variables, y = Valores, fill = Variables)) +
  geom_bar(stat = "identity") +
  labs(x = "Variables", y = "Media", title = "Gráfico de Barras de V1 y V2") +
  theme_minimal()
```

```r
# Crear un dataframe con los datos
data <- data.frame(
  Valores = c(97.0, 106.5, 111.0, 110.6, 115.0, 124.0,
              97.0, 107.5, 113.0, 111.8, 116.0, 128.0),
  Variable = rep(c("V1", "V2"), each = 6)
)
```

```r
# Cargar la librería necesaria
library(ggplot2)

# Crear el gráfico de caja
ggplot(data, aes(x = Variable, y = Valores, fill = Variable)) +
  geom_boxplot() +
  labs(x = "Variables", y = "Valores", title = "Gráfico de Caja de V1 y V2") +
  theme_minimal()
```

```r
# Crear un dataframe con los datos
data <- data.frame(
  V1 = c(97.0, 106.5, 111.0, 110.6, 115.0, 124.0),
  V2 = c(97.0, 107.5, 113.0, 111.8, 116.0, 128.0)
)
```

```r
# Cargar la librería necesaria
library(ggplot2)

# Crear el gráfico de dispersión
ggplot(data, aes(x = V1, y = V2)) +
  geom_point(color = "blue", size = 3) +
  labs(x = "V1", y = "V2", title = "Gráfico de Dispersión entre V1 y V2") +
  theme_minimal()
```

```r
# Crear un dataframe con los datos
data <- data.frame(
  Valores = c(97.0, 106.5, 111.0, 110.6, 115.0, 124.0,
              97.0, 107.5, 113.0, 111.8, 116.0, 128.0),
  Variable = rep(c("V1", "V2"), each = 6)
)
```

```r
# Cargar la librería necesaria
library(ggplot2)

# Crear el gráfico de densidad
ggplot(data, aes(x = Valores, fill = Variable)) +
  geom_density(alpha = 0.5) +
  labs(x = "Valores", y = "Densidad", title = "Gráfico de Densidad de V1 y V2") +
  theme_minimal()
```
